```python
#Change Data descriptions to streamline the description on some of the columns

df['MultipleLines']=df['MultipleLines'].replace(to_replace='.*No phone service*',value='No',regex=True)
df['OnlineSecurity']=df['OnlineSecurity'].replace(to_replace='.*No internet service*',value='No',regex=True)
df['OnlineBackup']=df['OnlineBackup'].replace(to_replace='.*No internet service*',value='No',regex=True)
df['DeviceProtection']=df['DeviceProtection'].replace(to_replace='.*No internet service*',value='No',regex=True)
df['TechSupport']=df['TechSupport'].replace(to_replace='.*No internet service*',value='No',regex=True)
df['StreamingTV']=df['StreamingTV'].replace(to_replace='.*No internet service*',value='No',regex=True)
df['StreamingMovies']=df['StreamingMovies'].replace(to_replace='.*No internet service*',value='No',regex=True)
```

```python
#Converting the data in column MonthlyCharges to 2 decimal places

##Convert TotalCharges column to numeric values
df['MonthlyCharges'] = pd.to_numeric(df['MonthlyCharges'], errors='coerce')
##Rounding each value to 2 decimal places
df['MonthlyCharges'] = df['MonthlyCharges'].apply(lambda x: round(x, 2) if not pd.isnull(x) else x)

print(df['MonthlyCharges'])
```

```python
#Converting the data in column TotalCharges to 2 decimal places

##Convert TotalCharges column to numeric values
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
##Rounding each value to 2 decimal places
df['TotalCharges'] = df['TotalCharges'].apply(lambda x: round(x, 2) if not pd.isnull(x) else x)

print (df['TotalCharges'])
```

```python
#Changing datatype of the columns

df['customerID']=df['customerID'].astype(object)
df['gender']=df['gender'].astype(bool)
df['Partner']=df['Partner'].astype(bool)
df['Dependents']=df['Dependents'].astype(bool)
df['PhoneService']=df['PhoneService'].astype(bool)
df['MultipleLines']=df['MultipleLines'].astype(bool)
df['InternetService']=df['InternetService'].astype(str)
df['OnlineSecurity']=df['OnlineSecurity'].astype(bool)
df['OnlineBackup']=df['OnlineBackup'].astype(bool)
df['DeviceProtection']=df['DeviceProtection'].astype(bool)
df['TechSupport']=df['TechSupport'].astype(bool)
df['StreamingTV']=df['StreamingTV'].astype(bool)
df['StreamingMovies']=df['StreamingMovies'].astype(bool)
df['PaperlessBilling']=df['PaperlessBilling'].astype(bool)
df['TotalCharges']=df['TotalCharges'].astype(float)
df['Churn']=df['Churn'].astype(bool)
```
